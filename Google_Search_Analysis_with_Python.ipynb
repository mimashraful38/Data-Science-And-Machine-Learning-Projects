```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq()
```

```python
trends.build_payload(kw_list=["NumPy tutorial"])
data = trends.interest_by_region()
data = data.sort_values(by="NumPy tutorial", ascending=False)
data = data.head(10)
print(data)
```

```python
data.reset_index().plot(x="geoName", y="NumPy tutorial", 
                        figsize=(20,15), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['NumPy tutorial'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(20, 15))
data['NumPy tutorial'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for NumPy tutorial', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
